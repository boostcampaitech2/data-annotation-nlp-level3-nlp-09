```python
df = pd.read_csv('q100 - q100.csv', encoding='utf-8')
df.head(5)
```

```python
import re, copy

def switch_tag(target_df):
    def _swtich(target_df, column):
        for idx, sentence in enumerate(target_df[column]):
            temp = copy.deepcopy(sentence)
            if '&lt;' in temp or '&gt;' in temp or '&quot;' in temp:
                temp = temp.replace('&lt;', '<')
                temp = temp.replace('&gt;', '>')
                temp = temp.replace('&quot;', '"')

                target_df[column][idx] = temp
                print(temp)
                print(target_df[column][idx])

    _swtich(target_df, 'sentence')
    _swtich(target_df, 'sentence_with_entity')
    return target_df

def remove_no_relation_type(target_df):
    def _remove(target_df, column):
        for idx, entity in enumerate(target_df[column]):
            # print(type(entity), entity)
            temp = copy.deepcopy(entity)
            if "'no'" in temp:
                temp = temp.replace("'no'", """''""")
                target_df[column][idx] = temp
                print(temp)
                print(target_df[column][idx])
                
    _remove(target_df, 'subject_entity')
    _remove(target_df, 'object_entity')

    return target_df

def check(target_df):
    not_swtich_cnt = 0
    for idx, sentence in enumerate(target_df['sentence']):
        if '&lt;' in sentence or '&gt;' in sentence or '&quot;' in sentence:
            not_swtich_cnt += 1
            
    print('not_swtich_cnt', not_swtich_cnt)

    def _remove(target_df, column):
        not_remove_cnt = 0
        for idx, entity in enumerate(target_df[column]):
            # print(type(entity), entity)
            temp = copy.deepcopy(entity)
            if "'no'" in temp:
                not_remove_cnt += 1
        print(f'{column}: not_remove_cnt {not_remove_cnt}')
                
    _remove(target_df, 'subject_entity')
    _remove(target_df, 'object_entity')

temp_df = copy.deepcopy(df)
temp_df = switch_tag(temp_df)

temp_df = remove_no_relation_type(temp_df)
check(temp_df)

temp_df.to_csv('q100_pp.csv', encoding='utf-8')
```
